### Artificial neural network

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("bank_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

#### Assigning values to variables X and y, not including customer id and surname as it doesn't affect the model in any way

```python
X = df.iloc[:, 3:13]
y = df.iloc[:, 13]
```

```python
X
```

```python
y
```

#### Geography and gender are categorical features and hence, needs to be converted into dummy variables

```python
df['Geography'].unique()
```

```python
states=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

#### dropping geography and gender as they are already been converted into dummy features

```python
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X
```

```python
X=pd.concat([X,states,gender],axis=1)
```

```python
X
```

#### Selecting test and train data, taking 20 percent as test data and 80 percent as train data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_test.count()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

#### Making the ANN now!

##### Importing keras libraries and packages

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

###### Initialising the ANN

```python
classifier = Sequential()
```

##### Adding the input layer and first hidden layer, adding the second hidden layer, adding the output layer

```python
classifier.add(Dense(activation="relu", input_dim=11, units=6, kernel_initializer="uniform"))
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

##### Compiling the ANN!

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.summary()
```

##### Fitting the ANN to training set

```python
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

##### Predicting the test set results

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

##### Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

#### We got an accuracy of 84.2 percent with this model
